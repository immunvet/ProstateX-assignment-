```python
values = t2_df['ClinSig'].value_counts().values
labels = t2_df['ClinSig'].value_counts().index

```

```python
#Function for plotting 

def plot_pie(data, column, name='', style = 'seaborn-pastel'):
    #shows absolute values count
    
    
    plt.style.use(style=style)
    
    values=data[column].value_counts().values
    lables=data[column].value_counts().index
    
    def absolute_value(val): #absolute value 
        a  = values[ np.abs(values - val/100.*values.sum()).argmin() ]
        return a
    
    explode=(0.0, 0.1)
    
    plt.pie(values, explode=explode, labels=labels, 
            autopct=absolute_value, 
            shadow=True, startangle=90)
    
    plt.axis('equal')
    
    plt.title(name)
    
    
```

```python
the_grid = GridSpec(2,2)

plt.subplot(the_grid[0,0], aspect=1)
plot_pie(t2_df, 'ClinSig', 't2 Tumor or Not')

plt.subplot(the_grid[0,1], aspect=1)
plot_pie(adc_df, 'ClinSig', 'adc Tumor or Not')

plt.subplot(the_grid[1,0], aspect=1)
plot_pie(bval_df, 'ClinSig', 'bval Tumor or Not')

plt.subplot(the_grid[1,1], aspect=1)
plot_pie(ktrans_df, 'ClinSig', 'ktrans Tumor or Not')


'''

The result that we have before data augmentation shows that we need to add more 
'cancer' images using data augmentation techniques
 
'''    
```

```python
'''
Our next step is to calculate the location of biopsy 

We have three different locations which might be used as an important feature for model training 

Plotting our location information 

'''
the_grid = GridSpec(2,2)

plt.subplot(the_grid[0,0], aspect=1)
plot_pie(t2_df, 'ClinSig', 't2 Biopsy Locations')

plt.subplot(the_grid[0,1], aspect=1)
plot_pie(adc_df, 'ClinSig', 'adc Biopsy Locations')

plt.subplot(the_grid[1,0], aspect=1)
plot_pie(bval_df, 'ClinSig', 'bval Biopsy Locations')

plt.subplot(the_grid[1,1], aspect=1)
plot_pie(ktrans_df, 'ClinSig', 'ktrans Biopsy Locations')
```

```python
#Function for plotting certain column from data frame 

def plot_clin_sig(data, column, name):
    plt.style.use('seaborn-notebook')
    plt.figure(figsize=(12,7))
    plt.suptitle(name, fontsize=20)
    ax=data['ClinSig'].value_counts().plot(kind='bar')
    ax.set_xlabel('{} Values` Names'.format(data.loc[:,column].name), fontsize='large')
    ax.set_ylabel('{} Values` Counts'.format(data.loc[:,column].name), fontsize='large')
    plt.show()


plot_clin_sig(t2_df, 'ClinSig', name='Cancer or Not?')
```
