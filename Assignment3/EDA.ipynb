```python
#First of all plotting ClinSigns for all of the modalities separately to get numbers of how many paties were diagnosed 
#with cancer after biopsy 
def plot_feature(data, column_name):
    #data=corresponding dataframe from which you want to plot a figure
    #column_name=column which you want to plot
    
    ax=data.loc[:,column_name].value_counts().plot(kind='bar')
    
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()
    
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
    
    plt.title('{} value counts'.format(column_name), fontsize=20)
    plt.xlabel('Number of {}'.format(column_name), fontsize=14)
    
    return plt.show()

   
#Plotting 

plt.subplots()


plt.subplot(1,2,1)
ax1=plot_feature(t2_df,'ClinSig')
plt.subplot(1,2,2)
ax2=plot_feature(adc_df,'ClinSig')
plt.subplot(2,2,1)
ax3=plot_feature(bval_df,'ClinSig')
plt.subplot(2,2,2)
ax4=plot_feature(ktrans_df,'ClinSig')
```
