```python
#reading data from prepared pickle dataframes

t2=pd.read_pickle('./t2_optimized.pkl')
adc=pd.read_pickle('./adc_optimized.pkl')
bval=pd.read_pickle('./bval_optimized.pkl')
ktrans=pd.read_pickle('./ktrans_optimized.pkl')

#reindexing data in random order
#I used this to increase distribution of ClinSig outcome among dataset, as this operation increased accuracy 
#for 2-5% percents

t2=t2.reindex(np.random.permutation(t2.index))
t2.index=range(len(t2))

adc=adc.reindex(np.random.permutation(adc.index))
adc.index=range(len(adc))

bval=bval.reindex(np.random.permutation(bval.index))
bval.index=range(len(bval))

ktrans=ktrans.reindex(np.random.permutation(ktrans.index))
ktrans.index=range(len(ktrans))
```

### Creating arrays for training and testing of our future models

```python

#creating arrays for training future models

def get_arrays(data, column=['patches', 'ClinSig', 'zone']):
    
    #get arrays from columns of dataframe
    #preserving perfect matching x to y, as well as saves correct shape of the array, which is not possible when using 
    #pandas
    
    x= np.array([i for i in data[column[0]]])#array of images
    y=np.array([i for i in data[column[1]]])#array for labels
    z=np.array([i for i in data[column[2]]])#array for zones
    return x,y,z

frames=[t2, adc, bval, ktrans]

t2_X, t2_Y, t2_zones = get_arrays(frames[0])
adc_X, adc_Y, adc_zones = get_arrays(frames[1])
bval_X, bval_Y, bval_zones = get_arrays(frames[2])
ktrans_X, ktrans_Y, ktrans_zones = get_arrays(frames[3])





```

## Zones to add them as location features

```python
#Getting Zones Arrays for each Modality
def get_zone_cat(data):
    encoder = LabelEncoder()
    encoder.fit(data)
    data = encoder.transform(data)
    #Getting Categorical Values
    data = keras.utils.to_categorical(data, 4)
    return data 


zones_frames=[t2_zones,adc_zones,bval_zones,ktrans_zones]


t2_Z = get_zone_cat(zones_frames[0])
adc_Z = get_zone_cat(zones_frames[1])
bval_Z = get_zone_cat(zones_frames[2])
ktrans_Z=get_zone_cat(zones_frames[3])
```

## Reshaping to tf format

```python
#reshape train
def reshaping(data=[], dimension = 1):

    #provide train and test datasets/arrays
    #provide dimension, which is 1 for gray colored images
    
    x_train = data.reshape(data.shape[0], data.shape[1], data.shape[2], dimension)
    
    
    return x_train

x_frames = [t2_X, adc_X, bval_X, ktrans_X]

t2_X=reshaping(x_frames[0])
adc_X=reshaping(x_frames[1])
bval_X=reshaping(x_frames[2])
ktrans_X=reshaping(x_frames[3])

```

## Labels Encoding 

```python
#labels encoder

def encode_labels(data):
    encoder=LabelEncoder()
    encoder.fit(data)
    encoded=encoder.transform(data)
    return encoded


labels_frames = [t2_Y, adc_Y, bval_Y, ktrans_Y]


t2_Y=encode_labels(labels_frames[0])
adc_Y=encode_labels(labels_frames[1])
bval_Y=encode_labels(labels_frames[2])
ktrans_Y=encode_labels(labels_frames[3])

```

# Training and Validation Part

## Model Constructing

```python
#compiling model
#defining model` function in order to use the same model for all modalities
#we use simple CNN model with adding location features to the last fully connected neuron

def gen_model(ToPrintOrNotToPrint=True):
    input_shape=(32,32,1)

    inp1=Input(shape=input_shape, name='inp1')
    inp2=Input(shape=(4,), name='zone')

    first_CN=Conv2D(32, kernel_size=(3,3))(inp1)
    act_1=Activation('relu')(first_CN)

    conv_2=Conv2D(64, (3,3))(act_1)
    act_2=Activation('relu')(conv_2)

    pool_1=MaxPooling2D(2,2)(act_2)
    conv_3=Conv2D(128, (3,3))(pool_1)
    act_3=Activation('relu')(conv_3)

    flat=Flatten()(act_3)
    fc1=Dense(192, activation='relu')(flat)


    merge=Concatenate()([fc1, inp2])

    fcn2=Dense(128, activation='relu')(merge)

    out=Dense(1, activation='sigmoid', name='out')(fcn2)

    model = Model(inputs=[inp1, inp2], outputs=out)
    
    
    if ToPrintOrNotToPrint==True:
        print(model.summary())
    
    
    return model
    
    


```

## Callbacks function

```python
#We can apply callbacks function to save only best results and maximizing training procedure 

def get_callbacks(name_weights, patience_lr):
    
    mcp_save = ModelCheckpoint(name_weights, save_best_only=True, monitor='val_loss', mode='min')
    
    reduce_lr_loss = ReduceLROnPlateau(monitor='loss', factor=0.1, patience=patience_lr, verbose=1, epsilon=1e-4, mode='min')
    
    return [mcp_save, reduce_lr_loss]

name_weights = "final_model_fold" + 'modality' + "_weights.h5"

callbacks = get_callbacks(name_weights = name_weights, patience_lr=10)
```

## t2 training 

```python
def mean_pred(y_true, y_pred):
    return K.mean(y_pred)


t2_model=gen_model()




t2_model.compile(loss=keras.losses.binary_crossentropy,  #using binary_crossentropy as we are solving binary classification problem
              optimizer=keras.optimizers.Adam(),
              metrics=['accuracy', mean_pred, metrics.mae])

t2_history=t2_model.fit({'inp1':t2_X, 'zone': t2_Z},
          {'out': t2_Y},
          epochs=10, batch_size=25,validation_split=0.25,shuffle=True, callbacks=callbacks)

```

```python
ktrans_model = gen_model()

ktrans_model.compile(loss=keras.losses.binary_crossentropy,  #using binary_crossentropy as we are solving binary classification problem
              optimizer=keras.optimizers.Adam(),
              metrics=['accuracy'])

ktrans_history=ktrans_model.fit({'inp1':ktrans_X, 'zone': ktrans_Z},
          {'out': ktrans_Y},
          epochs=10, batch_size=15,validation_split=0.25,shuffle=True, callbacks=callbacks)
```

```python
def plot_history(history):

    print(history.history.keys())

    # summarize history for accuracy
    plt.plot(history.history['acc'])
    plt.plot(history.history['val_acc'])
    plt.title('model accuracy')
    plt.ylabel('accuracy')
    plt.xlabel('epoch')
    plt.legend(['train', 'test'], loc='upper left')
    plt.show()
    # summarize history for loss
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title('model loss')
    plt.ylabel('loss')
    plt.xlabel('epoch')
    plt.legend(['train', 'test'], loc='upper left')
    plt.show()
```
