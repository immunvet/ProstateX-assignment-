```python
#Reading all of the .csv files and creating a dataframes at first
t2_df=pd.read_csv('./t2_df.csv')
adc_df=pd.read_csv('./adc_df.csv')
bval_df=pd.read_csv('./bval_df.csv')
ktrans_df=pd.read_csv('./ktrans_df.csv')
```

```python
#Function for plotting 

def plot_pie(data, column, name='', style = 'seaborn-pastel'):
    #shows absolute values count
    
    
    plt.style.use(style=style)
    
    values=data[column].value_counts().values
    labels=data[column].value_counts().index
    
    def absolute_value(val): #absolute value 
        a  = values[ np.abs(values - val/100.*values.sum()).argmin() ]
        return a
    
    explode=tuple([0.1 if values[i]==values.max() else 0.0 for i in range(len(values))]) #defining explode for maximum value
    
    
    plt.pie(values, labels=labels, 
            explode=explode,
            autopct=absolute_value, 
            shadow=True, startangle=90)
    
    plt.axis('equal')
    
    plt.title(name)
    
    
```

```python
'''
The result that we have before data augmentation shows that we need to add more 
cancer images using data augmentation techniques 
'''    


the_grid = GridSpec(2,2)

plt.subplot(the_grid[0,0], aspect=1)
plot_pie(t2_df, 'ClinSig', 't2 Tumor or Not')

plt.subplot(the_grid[0,1], aspect=1)
plot_pie(adc_df, 'ClinSig', 'adc Tumor or Not')

plt.subplot(the_grid[1,0], aspect=1)
plot_pie(bval_df, 'ClinSig', 'bval Tumor or Not')

plt.subplot(the_grid[1,1], aspect=1)
plot_pie(ktrans_df, 'ClinSig', 'ktrans Tumor or Not')


```

```python
'''
Our next step is to calculate the location of biopsy 

We have three different locations which might be used as an important feature for model training 

Plotting our location information 

'''
the_grid = GridSpec(2,2)

plt.subplot(the_grid[0,0], aspect=1)
plot_pie(t2_df, 'zone', 't2 Biopsy Locations', style ='seaborn-paper')

plt.subplot(the_grid[0,1], aspect=1)
plot_pie(adc_df, 'zone', 'adc Biopsy Locations', style ='seaborn-paper')

plt.subplot(the_grid[1,0], aspect=1)
plot_pie(bval_df, 'zone', 'bval Biopsy Locations', style ='seaborn-paper')

plt.subplot(the_grid[1,1], aspect=1)
plot_pie(ktrans_df, 'zone', 'ktrans Biopsy Locations', style ='seaborn-paper')



```
